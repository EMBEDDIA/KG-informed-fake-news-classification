```python
import pandas as pd
import time
import pickle
```

```python
import wikidata_concept_finder
```

```python
datasets = ["pan2020","LIAR_PANTS", "AAAI2021_COVID19_fake_news", "FakeNewsNet", "ISOT"]
```

```python
for dataset in ["LIAR_PANTS", "AAAI2021_COVID19_fake_news", "FakeNewsNet", "ISOT"]:
    split_path = "data/final/"+dataset
    for t in ["train", "test", "dev"]:       
        data = pd.read_csv(split_path+"/"+t+".csv")
        train_text = data["text_a"].to_list()
        y = list(train_text)
        for kg in ["complex", "transe", "quate", "simple", "rotate", "distmult"]: # 
            start = time.time()
            ys = wikidata_concept_finder.naive_embedd(y, 3, kg)
            end = time.time()
            print(dataset, t, len(y), "TIME NEEDED: ", end - start)
            out_path = "kg_emb_dump/"+dataset+"/"+t+"_"+kg+"_n.pkl"
            with open(out_path, "wb") as f:
                pickle.dump(ys, f)
            print("DATA DROPPED @ ", out_path)
```
