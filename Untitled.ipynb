```python
import pandas as pd
from tqdm import tqdm
import numpy as np
datasets = [  "pan2020",  "LIAR_PANTS", "ISOT", "AAAI2021_COVID19_fake_news", "FakeNewsNet"]
for dataset in tqdm(datasets):
    path = "representations/"+dataset+"/"
    for thing in  tqdm(["train", "dev", "test"]):
        path_in = "data/final/"+dataset+"/"+thing+'.csv'
        df = pd.read_csv(path_in, encoding='utf-8')
        texts = df.text_a.to_list()
        name = dataset + "_" + thing
        matrixs = lsa_features.fit_space(texts,nfeat = 5000, dim = 512, split = thing, name = dataset  )
        path_tmp = path + thing+"/lsa.csv"
        np.savetxt(path_tmp, matrixs, delimiter=",")
```
